# Model Preparation  
  
**MODEL CHOICE: Support Vector Machine**  
  
> **Definition & Explanation:**  
>  
> In machine learning, support vector machines (SVMs, also support vector networks) are supervised learning models with associated learning algorithms that analyze data used for classification and regression analysis. Given a set of training examples, each marked as belonging to one or the other of two categories, an SVM training algorithm builds a model that assigns new examples to one category or the other, making it a non-probabilistic binary linear classifier (although methods such as Platt scaling exist to use SVM in a probabilistic classification setting). An SVM model is a representation of the examples as points in space, mapped so that the examples of the separate categories are divided by a clear gap that is as wide as possible. New examples are then mapped into that same space and predicted to belong to a category based on which side of the gap they fall.
  
> In addition to performing linear classification, SVMs can efficiently perform a non-linear classification using what is called the kernel trick, implicitly mapping their inputs into high-dimensional feature spaces.
  
**Source:** [https://en.wikipedia.org/wiki/Support_vector_machine](https://en.wikipedia.org/wiki/Support_vector_machine)

```python
### CHANGE PASS / FAIL COLUMN TO CATEGORICAL DATA TYPE ###

df['p_f'] = df.p_f.astype('category')
```

```python
### SPLIT THE DATA INTO TRAINING & TESTING SUBSETS ###

X = df.loc[:, ['f2', 'f3', 'f4', 'f6', 'f8']]
y = df.loc[:, 'p_f']

train_x, test_x, train_y, test_y = train_test_split(X, y, test_size = 0.20)
```

```python
### SCALE THE FEATURES ###

# Standardize features by removing the mean & scaling to unit variance: 
scale = StandardScaler()
scale.fit(train_x)

x_train = scale.transform(train_x)
x_test = scale.transform(test_x)
# Scaling helps to keep the variables with larger numbers & greater variance 
# from "overpowering" the model and giving themselves more weight.  
```

```python
### TRAIN & TEST SVM MODEL ###

model = svm.SVC(C=1, kernel='poly', degree=3)
y_hat = model.fit(x_train, train_y).predict(x_test)
print("Number of mislabeled points out of a total %d points : %d"
      % (test_y.shape[0],(test_y != y_hat).sum()))
```

```python
### VISUALIZE MODEL RESULTS IN A CONTINGENCY TABLE ### 

skplt.metrics.plot_confusion_matrix(test_y, y_hat, normalize=True);
```

### Some definitions:  
  
 - Recall is the number of True Positives divided by the number of True Positives and the number of False Negatives.  
 - Precision is the number of True Positives divided by the number of True Positives and False Positives.  
 - The F$_1$ Score is the 2\*((precision \* recall)/(precision + recall)).
 
***Source:*** [https://machinelearningmastery.com/](https://machinelearningmastery.com/classification-accuracy-is-not-enough-more-performance-measures-you-can-use/)

```python
### CREATE A BASIC CONFUSION MATRIX TO REFERENCE FOR CALCULATIONS ###

conf = confusion_matrix(test_y, y_hat)
conf
```

```python
### CALCULATE RECALL, PRECISION & F1 SCORE ###

recall = conf[0, 0]/(conf[0, 0] + conf[1, 0])
precision = conf[0, 0]/(conf[0, 0] + conf[0, 1])
f1_score = 2 * ((precision * recall) / (precision + recall))

print('Recall:    ', round(recall, 3))
print('Precision: ', round(precision, 3))
print('F1_Score:  ', round(f1_score, 3))
```

```python
### PLOT FEATURE IMPORTANCE IN THE MODEL ###

from sklearn import svm
svm = svm.SVC(kernel='linear').fit(x_train, train_y)
sns.barplot(x=X.columns.values, y=abs(svm.coef_[0]))
plt.title('Model Input Feature ImportanceIn Absolute Values, for Comparison')
plt.xlabel('Feature Name')
plt.ylabel('Model Coefficient Output\nIn Absolute Values');
```

```python
print('`f4` is the feature that contributes the most to the model output.  Coefficient:', svm.coef_.max())
print('`f8` is the feature that contributes the second-most to the model output.  Coefficient:', svm.coef_[0][-1:])
```
